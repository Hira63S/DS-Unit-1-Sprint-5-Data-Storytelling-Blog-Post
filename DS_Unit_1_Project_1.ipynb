```
subset = (under25[['Item', 'Housing']]
           .merge(over25_34[['Item', 'Housing']]))                 
```

```
import matplotlib.ticker as ticker

ax = subset.plot.bar(color=('#47d1d1', '#FF7633'), width=0.9);

ax.set(xlabel='Rating', ylabel='Percent of total percent',
      yticks=range(0,70,10), facecolor='#f9f9f9')

#formatter = ticker.FormatStrFormatter('%%')
#ax.yaxis.set_major_formatter(formatter)

#axes.text is very flexible, 
ax.text(x=-2, y=75, s="An inconvenient Sequal, Truth to Power", fontsize = 14,
       fontweight='bold');

ax.text(x=-2, y=72, s='IMDB ratings for the film as of Aug.29', fontsize=12);
```
